### Get ResNet50 Features - 

```python
from keras.applications.resnet50 import ResNet50, preprocess_input
from keras.preprocessing import image
import numpy as np
import os
import re

base_model =  ResNet50(weights='imagenet', include_top=True)
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc1000').output)

def extract_features(img_paths, batch_size=64):
    """ This function extracts image features for each image in img_paths using ResNet50 bottleneck layer.
        Returned features is a numpy array with shape (len(img_paths), 2048).
    """
    global model
    n = len(img_paths)
    img_array = np.zeros((n, 224, 224, 3))
    
    for i, path in enumerate(img_paths):
        img = image.load_img(path, target_size=(224, 224))
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis=0)
        x = preprocess_input(img)
        img_array[i] = x
    
    X = model.predict(img_array, batch_size=batch_size, verbose=1)
    X = X.reshape((n, -1))
    return X
```

```python
##  Type1
images_dir = '../train_bkp/Type_1/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
RN_train_type1=extract_features(list_images)
##  Type2
images_dir = '../train_bkp/Type_2/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
RN_train_type2=extract_features(list_images)
##  Type3
images_dir = '../train_bkp/Type_3/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
RN_train_type3=extract_features(list_images)
# Test
images_dir = '../test/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
RN_test=extract_features(list_images)
```

```python
import pandas as pd
#save results

RN_train_type1= pd.DataFrame(RN_train_type1)
RN_train_type2= pd.DataFrame(RN_train_type2)
RN_train_type1.append(RN_train_type2)
RN_train_type3= pd.DataFrame(RN_train_type3)
RN_train_type1.append(RN_train_type3)
RN_train_type1.to_csv("ResNet50_train_features.csv", index=False)

RN_test= pd.DataFrame(RN_test)
RN_test.to_csv("ResNet50_test_features.csv", index=False)
```

### Get Inceptionv3

```python
#https://github.com/fchollet/keras/blob/master/keras/applications/inception_v3.py

from keras.applications.inception_v3 import InceptionV3, preprocess_input
from keras.preprocessing import image
from keras.models import Model
import numpy as np
import os
import re

base_model = InceptionV3(weights='imagenet', include_top=True)
model = Model(inputs=base_model.input, outputs=base_model.get_layer('predictions').output)

def extract_features(img_paths, batch_size=64):
    """ This function extracts image features for each image in img_paths using ResNet50 bottleneck layer.
        Returned features is a numpy array with shape (len(img_paths), 2048).
    """
    global model
    n = len(img_paths)
    img_array = np.zeros((n, 299, 299, 3))
    
    for i, path in enumerate(img_paths):
        img = image.load_img(path, target_size=(299, 299))
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis=0)
        x = preprocess_input(img)
        img_array[i] = x
    
    X = model.predict(img_array, batch_size=batch_size, verbose=1)
    X = X.reshape((n, -1))
    return X
```

```python
##  Type1
images_dir = '../train_bkp/Type_1/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
Inception3_train_type1=extract_features(list_images)
##  Type2
images_dir = '../train_bkp/Type_2/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
Inception3_train_type2=extract_features(list_images)
##  Type3
images_dir = '../train_bkp/Type_3/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
Inception3_train_type3=extract_features(list_images)
# Test
images_dir = '../test/'
list_images = [images_dir+'/'+f for f in os.listdir(images_dir) if re.search('jpg|JPG', f)]
Inception3_test=extract_features(list_images)
```

```python
import pandas as pd
#save results

Inception3_train_type1= pd.DataFrame(Inception3_train_type1)
Inception3_train_type2= pd.DataFrame(Inception3_train_type2)
Inception3_train_type1.append(Inception3_train_type2)
Inception3_train_type3= pd.DataFrame(Inception3_train_type3)
Inception3_train_type1.append(Inception3_train_type3)
Inception3_train_type1.to_csv("Inception3_train_features.csv", index=False)

Inception3_test= pd.DataFrame(Inception3_test)
Inception3_test.to_csv("Inception3_test_features.csv", index=False)
```
